### Data wrangling

```python
# # filter out pieces that are not "in place"
# public_art_df = public_art_df.query('Status=="In place"')
```

```python
# drop nulls for "geopoint"
public_art_df.dropna(subset=['geo_point_2d'], inplace=True)
```

```python
public_art_df
```

```python
# filter out columns 
public_art_df = public_art_df[[
    'Title of Work', 'Type', 'Status', 
    'SiteAddress', 'URL', 'PhotoURL', 
    'Neighbourhood', 'DescriptionOfwork', 
    'YearOfInstallation', 'geo_point_2d'
]]
```

```python
public_art_df.info()
```

```python
# fill nulls with empty strings for "SiteAddress" and "DescriptionOfwork"
public_art_df.SiteAddress = public_art_df.SiteAddress.fillna('')
public_art_df.DescriptionOfwork = public_art_df.DescriptionOfwork.fillna('')
```

```python
public_art_df.info()
```

**Notes:**
- `PhotoURL`: can contain NaNs - the result is that the corresponding point on the map will not have an image
- `Neighbourhood`: missing values will be imputed using the latitude and longitude in the app script
